```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# Load dataset
df = pd.read_csv(r"C:\Users\Admin\Documents\global-space-analytics\data\database.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Top countries by active satellites 
top_countries = df['Country of Operator/Owner'].value_counts().head(10)
px.bar(
    top_countries,
    title="Top 10 Countries by Active Satellites",
    labels={'value':'Satellite Count','index':'Country'}
).show()
```

```python
# Distribution of Satellites by orbit type

px.pie(
    df,
    names='Class of Orbit',
    title='Distribution of Satellites by Orbit Type'
).show()
```

```python
# Purpose of Satellites in Space

purpose_counts = df['Purpose'].value_counts().head(10)
px.bar(
    purpose_counts,
    title="Top Satellite Purpose",
    labels={'value':'Count','Index':'Purpose'}
).show()
```
